```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('tetrahymena_padel_withDep.csv')
dataf1 = dataset.drop(['Name'], axis = 1)

dataf1.shape
```

```python
dataset_rmNA = dataf1.dropna(axis = 'columns', how = 'any')

dataset_rmNA.shape
```

```python
from sklearn.feature_selection import VarianceThreshold

threshold_n=0.90
sel = VarianceThreshold(threshold=(threshold_n* (1 - threshold_n) ))
sel_var=sel.fit_transform(dataset_rmNA)
dataset_VT = dataset_rmNA[dataset_rmNA.columns[sel.get_support(indices=True)]]
```

```python
dataset_VT_rmDep = dataset_VT.drop(['Dependent'], axis = 1)

corrWithDep = dataset_VT.corr().abs().iloc[0]
corrWithVar = dataset_VT_rmDep.corr().abs()
```

```python
col_corr = set()

for i in range(len(corrWithVar.columns)):
    for j in range(i):
        if (corrWithVar.iloc[i, j] >= 0.9) and (corrWithVar.columns[i] or corrWithVar.columns[j] not in col_corr):
            if (corrWithDep.iloc[i] >= corrWithDep.iloc[j]):
                colname = corrWithVar.columns[j]
                col_corr.add(colname)
            else:
                colname = corrWithVar.columns[i]
                col_corr.add(colname)
```

```python
dataset_PW = dataset_VT.drop(col_corr, axis = 1)
```

```python
dataset_cleaned = pd.concat([dataset['Name'], dataset_PW], axis = 1)
```
